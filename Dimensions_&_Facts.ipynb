1. DIM_Inventory_Location

```python
import pandas as pd
import random
from datetime import datetime, timedelta

rows = []

storage_types = ["Rack", "Shelf", "Cold Storage", "Bulk Storage"]

for i in range(1, 1001):

    location_type = random.choice(["Warehouse", "Store"])

    warehouse_id = f"W{random.randint(1,50)}"
    store_id = f"S{random.randint(1,200)}"

    # Storage type
    storage_type = random.choice(storage_types)

    # Cold storage flag logic
    if storage_type == "Cold Storage":
        cold_flag = "Yes"
    else:
        cold_flag = "No"

    # Stock level
    stock_level = random.randint(0, 500)

    # Inventory status based on stock level
    if stock_level == 0:
        status = "Out of Stock"
    elif stock_level < 50:
        status = "Low Stock"
    else:
        status = "In Stock"

    row = {
        "Inventory_Location_ID": f"IL{i}",
        "Location_Type": location_type,
        "Warehouse_ID": warehouse_id,
        "Store_ID": store_id,
        "Storage_Type": storage_type,
        "Cold_Storage_Flag": cold_flag,
        "SKU_ID": f"SKU{random.randint(1000,1999)}",
        "Stock_Level": stock_level,
        "Inventory_Status": status,
        "Last_Updated": (datetime.now() - timedelta(days=random.randint(0,30))).strftime("%Y-%m-%d")
    }

    rows.append(row)

df = pd.DataFrame(rows)

df.to_csv("DIM_Inventory_Location.csv", index=False)

print("DIM_Inventory_Location.csv")
```

ii. Fact_Inventory_Stock

```python
rows = []

for i in range(1,1001):

    on_hand = random.randint(50,500)

    reserved = random.randint(0, int(on_hand * 0.4))

    available = on_hand - reserved

    safety_stock = random.randint(20,100)

    reorder_point = safety_stock + random.randint(10,50)

    if available <= safety_stock:
        status = "Critical"
    elif available <= reorder_point:
        status = "Low"
    else:
        status = "Normal"

    row = {
        "Stock_Record_ID": f"SR{i}",
        "SKU_ID": f"SKU{random.randint(1,1000)}",
        "Inventory_Location_ID": f"IL{random.randint(1,500)}",
        "Time_ID": f"T{random.randint(1,1000)}",
        "On_Hand_Quantity": on_hand,
        "Reserved_Quantity": reserved,
        "Available_Quantity": available,
        "Safety_Stock_Level": safety_stock,
        "Reorder_Point": reorder_point,
        "Stock_Status": status
    }

    rows.append(row)

df = pd.DataFrame(rows)

df.to_csv("Fact_Inventory_Stock.csv", index=False)

print("Fact_Inventory_Stock.csv")
```
